```python
import thread as tH
import time as t
print "start time  : ",t.time()
def fun(name = None):
    for i in range(10):
        print name
        t.sleep(1)
def main():
    print "main process started "
    tH.start_new_thread(fun,("anjan",))
    print "main process is ended "
if __name__ =="__main__":
    main()
print "stop time  : ",t.time()
```

```python
import thread as tH
import time as t
print "start time  : ",t.time()
def fun1(name = None):
    for i in range(10):
        print name
        t.sleep(1)
def fun2(name = None):
    for i in range(10):
        print name
        t.sleep(1)
def fun3(name = None):
    for i in range(10):
        print name
        t.sleep(1)
def main():
    print "main process started "
    tH.start_new_thread(fun1,("anjan",))
    tH.start_new_thread(fun2,("kumar",))
    tH.start_new_thread(fun3,("python",))
    print "main process is ended "
if __name__ =="__main__":
    main()
print "stop time  : ",t.time()
```

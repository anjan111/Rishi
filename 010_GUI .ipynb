```python
# tkinter  ---> gui lib  ,  pyqt

import  tkinter as tkn
'''
Tk() --->> to create window object
Label() --->> Patiations of your windows
title() ---->> To title of window
mainloop() --> Untill we closed the windows , the window can visible 
'''
window = tkn.Tk()
window.title("This is 1st GUI Application")

lb1 = tkn.Label(window,text = "This is Label 1")
lb2 = tkn.Label(window,text = "This is Label 2")
lb3 = tkn.Label(window,text = "This is Label 3")
lb4 = tkn.Label(window,text = "This is Label 4")
lb1.grid(column =0 ,row = 0)
lb2.grid(column =1 ,row = 0)
lb3.grid(column =0 ,row = 1)
lb4.grid(column =1 ,row = 1)

window.mainloop()
```

```python
# tkinter  ---> gui lib  ,  pyqt

import  tkinter as tkn
def task1():
    lb1.config(text = "You Pressed Button1")
    lb2.config(text = " ")
    lb3.config(text = " ")
    lb4.config(text = " ")
def task2():
    lb2.config(text = "You Pressed Button2")
    lb1.config(text = " ")
    lb3.config(text = " ")
    lb4.config(text = " ")
def task3():
    lb3.config(text = "You Pressed Button3")
    lb2.config(text = " ")
    lb1.config(text = " ")
    lb4.config(text = " ")
def task4():
    lb4.config(text = "You Pressed Button4")
    lb2.config(text = " ")
    lb3.config(text = " ")
    lb1.config(text = " ")

window = tkn.Tk()
window.title("This is 1st GUI Application")

lb1 = tkn.Label(window,text = "This is Label 1")
lb2 = tkn.Label(window,text = "This is Label 2")
lb3 = tkn.Label(window,text = "This is Label 3")
lb4 = tkn.Label(window,text = "This is Label 4")

btn1  = tkn.Button(window , text = "1", command = task1)
btn2  = tkn.Button(window , text = "2", command = task2)
btn3  = tkn.Button(window , text = "3", command = task3 )
btn4  = tkn.Button(window , text = "4", command = task4)

lb1.grid(column =0 ,row = 0)
lb2.grid(column =1 ,row = 0)
lb3.grid(column =0 ,row = 1)
lb4.grid(column =1 ,row = 1)

btn1.grid(column =0 ,row = 2)
btn2.grid(column =1 ,row = 2)
btn3.grid(column =0 ,row = 3)
btn4.grid(column =1 ,row = 3)

window.mainloop()
```

```python
# tkinter  ---> gui lib  ,  pyqt

import  tkinter as tkn
def task1():
    Text = "entered  data is "+ent.get()
    lb2.config(text = Text)
    
window = tkn.Tk()
window.title("This is 1st GUI Application")

lb1 = tkn.Label(window,text = "Enter some text : ")
lb2 = tkn.Label(window,text = "This is Label 2")
btn1  = tkn.Button(window , text = "Click me", command = task1)

ent =  tkn.Entry(window)

lb1.grid(column =0 ,row = 0)
ent.grid(column =1 ,row = 0)
btn1.grid(column =0 ,row = 1)
lb2.grid(column = 0 ,row = 2)

window.mainloop()
```
